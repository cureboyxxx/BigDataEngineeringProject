Ethical web scraping involves respecting website terms of service, avoiding overloading servers, and ensuring that the collected data is used responsibly and in compliance with privacy laws and regulations.

Use Python, ```requests```, ```BeautifulSoup``` and/or ```pandas``` to scrape web data:

## Import Libraries

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

## Define the Target URL

````python
url = "https://en.wikipedia.org/wiki/IPhone"
````

## Send a Request to the Website

Do not forget to check the response status code

````python
response = requests.get(url)
print(f"Status Code: {response.status_code}")
````

````python
# Well, then we will use a User-Agent
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) TechShop_UniProject_Scraper/1.0"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    print("Request successful! Status Code: 200")
else:
    print(f"Error: {response.status_code}")
````

## Parse the HTML Content

Use a library to access the HTMl content

````python
soup = BeautifulSoup(response.text, 'html.parser')
print("HTML parsed succesfully!")
````

## Identify the Data to be Scraped

I want to scrape the complete list of Apple iPhone models along with their release dates, technical specifications, and support lifespans. This information is perfectly structured in the second table of the Wikipedia page (under the "Models" section). Extracting this specific table will give us a clean dataset of all iPhone generations to analyze their evolution.

## Extract Data

Find specific elements and extract text or attributes from elements (handle pagination if necessary)

````python
# Find every Wikipedia tables
tables = soup.find_all('table', class_='wikitable')

# Choose the second one from them
target_table = tables[1]
print("Table found!")
````

## Store Data in a Structured Format

Give a brief overview of the data collected (e.g. count, fields, ...)

````python
# Pandas converts the HTML-Code of the table directly to a DataFrame
df = pd.read_html(str(target_table))[0]

# Clean up: Drop the last 4 rows because they contain the legend and remarks, not actual iPhone models.
df = df.iloc[:-4]

print(f"Found: {df.shape[0]} rows and {df.shape[1]} columns.")
display(df.head())
````

## Save the Data

````python
df.to_csv("wikipedia_iphone_models.csv", index=False)
print("Data saved!")
````
